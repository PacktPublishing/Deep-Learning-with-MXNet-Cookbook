```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
house_df = pd.read_csv("kc_house_data.csv")
```

```python
house_df.info()
```

```python
house_df.head()
```

```python
house_df.hist(column = "price", bins = 24)
plt.show()
```

```python
house_df = house_df.drop(["id"], axis=1)
house_corr = house_df.corr()
```

```python
plt.figure(figsize=(20, 20))
colormap = sns.color_palette("rocket_r", as_cmap=True)
sns.heatmap(house_corr, annot=True, cmap=colormap)
plt.show()
```

```python
house_corr["price"].drop(["price"]).sort_values(ascending = False).plot.bar(figsize=(5,5))
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['sqft_living'], house_df['price'], alpha = 0.1)
plt.xlabel("Living Square Feet")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['sqft_above'], house_df['price'], alpha = 0.1)
plt.xlabel("Above Square Feet")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['sqft_living15'], house_df['price'], alpha = 0.1)
plt.xlabel("Neighbors Living Square Feet")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['grade'], house_df['price'], alpha = 0.1)
plt.xlabel("Grade")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['floors'], house_df['price'], alpha = 0.1)
plt.xlabel("Number of Floors")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df["bedrooms"], house_df["price"], alpha = 0.1)
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['bathrooms'], house_df['price'], alpha = 0.1)
plt.xlabel("Number of Bathrooms")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['view'], house_df['price'], alpha = 0.1)
plt.xticks(range(0, 5))
plt.xlabel("View Quality")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df['waterfront'], house_df['price'], alpha = 0.1)
plt.xticks([-1, 0, 1, 2], ["", "Without", "With", ""])
plt.xlabel("Waterfront View")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df["yr_built"], house_df["price"], alpha = 0.1)
plt.xlabel("Construction Year")
plt.ylabel("Price ($)")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(house_df["yr_renovated"] != 0, house_df["price"], alpha = 0.1)
plt.xticks([-1, 0, 1, 2], ["", "Not Renovated", "Renovated", ""])
plt.xlabel("Renovated Houses")
plt.ylabel("Price ($)")
plt.show()
```

```python
house_df.plot.scatter(x="long", y="lat", c="price",
              #cmap=plt.get_cmap("gist_rainbow"), colorbar=True,
              cmap=plt.get_cmap("spectral"), colorbar=True,
              alpha=0.1, figsize=(20,20))
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
house_df.plot.scatter(x="long", y="lat", c="price",
              cmap=plt.get_cmap("cubehelix_r"), colorbar=True,
              alpha=0.1, figsize=(20,20))
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
