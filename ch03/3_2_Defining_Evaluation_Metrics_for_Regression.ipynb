```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Input Data
difference = np.arange(-5, 5, 0.01)
```

```python
# Mean Absolute Error
MAE = abs(difference)
```

```python
plt.plot(difference, MAE)
plt.xlabel("Mean Difference")
plt.ylabel("Mean Absolute Error")
plt.show()
```

```python
# Mean Squared Error
MSE = difference**2
```

```python
plt.plot(difference, MSE)
plt.xlabel("Mean Difference")
plt.ylabel("Mean Squared Error")
plt.show()
```

```python
# Smooth Mean Absolute Error
def smooth_MAE(difference):
    return (abs(difference) >= 1.0) * (abs(difference) - 0.5) + (abs(difference) < 1.0) * 0.5 * difference**2
```

```python
plt.plot(difference, smooth_MAE(difference))
plt.xlabel("Mean Difference")
plt.ylabel("Smooth Mean Absolute Error")
plt.show()
```
